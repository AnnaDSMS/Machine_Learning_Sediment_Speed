```python
#read biovolume measurements
r_filenameTSV = '/Users/annasommer/Documents/Documents/Norwich_Project/Data/uvp_tara/size_spectra.csv'
tsv_read_columns = pd.read_csv(r_filenameTSV)
```

```python
biovol_mm = tsv_read_columns[['part_sampleid', 'depth', 'watervolume', 'biovol18', 'biovol19',\
'biovol20', 'biovol21', 'biovol22', 'biovol23', 'biovol24', 'biovol25',\
'biovol26', 'biovol27', 'biovol28', 'biovol29', 'biovol30', 'biovol31',\
'biovol32', 'biovol33', 'biovol34', 'biovol35', 'biovol36', 'biovol37',\
'biovol38', 'biovol39', 'biovol40', 'biovol41', 'biovol42', 'biovol43',\
'biovol44', 'biovol45']]
```

```python
#convert mm3/L to mm3
for i in np.arange(0,tsv_read_columns.columns.size,1):
    name_biovol = tsv_read_columns.columns[i]  
    if name_biovol[:6] == 'biovol':
        biovol_mm[name_biovol] = biovol_mm[name_biovol]
```

```python
biovol_mm
```

```python
biomass_estim = biovol_mm[['part_sampleid', 'depth', 'watervolume', 'biovol18', 'biovol19',\
'biovol20', 'biovol21', 'biovol22', 'biovol23', 'biovol24', 'biovol25',\
'biovol26', 'biovol27', 'biovol28', 'biovol29', 'biovol30', 'biovol31',\
'biovol32', 'biovol33', 'biovol34', 'biovol35', 'biovol36', 'biovol37',\
'biovol38', 'biovol39', 'biovol40', 'biovol41', 'biovol42', 'biovol43',\
'biovol44', 'biovol45']]
```

```python
#convert to biomass accordong to Alldredge
for i in np.arange(0,tsv_read_columns.columns.size,1):
    name_biovol = tsv_read_columns.columns[i]  
    if name_biovol[:6] == 'biovol':
        biomass_estim[name_biovol] = 3.33 * biomass_estim[name_biovol]**0.54
```

```python
biomass_estim
```

```python
biomass_estim_mmL = biomass_estim[['part_sampleid', 'depth', 'watervolume', 'biovol18', 'biovol19',\
'biovol20', 'biovol21', 'biovol22', 'biovol23', 'biovol24', 'biovol25',\
'biovol26', 'biovol27', 'biovol28', 'biovol29', 'biovol30', 'biovol31',\
'biovol32', 'biovol33', 'biovol34', 'biovol35', 'biovol36', 'biovol37',\
'biovol38', 'biovol39', 'biovol40', 'biovol41', 'biovol42', 'biovol43',\
'biovol44', 'biovol45']]
```

```python
#biomass per L
for i in np.arange(0,tsv_read_columns.columns.size,1):
    name_biovol = tsv_read_columns.columns[i]  
    if name_biovol[:6] == 'biovol':
        biomass_estim_mmL[name_biovol] = biomass_estim[name_biovol] / biomass_estim['watervolume']
```
